```python
import pandas as pd
def mridataorg_to_pandas(url= 'http://mridata.org/list?project=NYU%20machine%20learning%20data'):
    tables = pd.read_html(url)
    tables = [t.set_index(0).T for t in tables if 
              (len(t.columns) == 2 and any(t.iloc[:,0].str.contains('UUID')))]
    tables = pd.concat(tables)
    return tables
def mridataorgfile_to_pandas(htmlFile='mridata.org-100.html'):
    tables = pd.read_html(htmlFile, flavor='bs4') #I think it accepts BeatifulSoup object
    tables = [t.set_index(0).T for t in tables if 
                (len(t.columns) == 2 and any(t.iloc[:,0].str.contains('UUID')))]
    tables = pd.concat(tables)
    return tables
```

```python
all_tables = mridataorgfile_to_pandas()
tables = all_tables.loc[all_tables['Trajectory'] == 'cartesian']
tables
```
